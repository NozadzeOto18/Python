# (L6 – 3 ქულა) გამოიყენეთ მონაცემთა წმედის რომელიმე ფუნქციონალი (მაგ. რომელიმე სვეტის მონაცემები გადაანაწილეთ რამდენიმე ახალ სვეტად, წაშალეთ რომელიმე სვეტი, ცარიელი მნიშვნელობებისთვი გამოიყენეთ სასურველი ფორმატი, ა.შ.)

```python
import pandas as pd
df = pd.read_csv("Earthquakes_database.csv")
df = df.drop_duplicates()
df
```

```python
df = df.drop(columns = "Type")
df
```

```python
df[["Month", "Day", "Year"]] = df["Date"].str.split('/',2, expand=True)
df
```

```python
df["Status"] = df["Status"].str.replace('Automatic','Auto')
df
```

# (L6 – 3 ქულა) გამოიყენეთ მონაცემთა წმედის რომელიმე ფუნქციონალი (მაგ. რომელიმე სვეტის მონაცემები გააანაწილეთ რამდენიმე ახალ სვეტად, წაშალეთ რომელიმე სვეტი, ცარიელი მნიშვნელობებისთვი 
გამოიყენეთ სასურველი ფორმატი, ა.შ.)

```python
#df.sort_values(by='Magnitude', ascending=True)
df['Power'] = pd.cut(df['Magnitude'], 
       bins=[int(df['Magnitude'].min())-1, 5.5, 9.1, int(df['Magnitude'].max())+1],
       labels=['low', 'medium', 'high'])
df
```

# (L10 - 3 ქულა) ააგეთ თქვენთვის სასურველი 4 განსხვავებული დიაგრამა შემდეგი ტიპებიდან (ხაზოვანი დიაგრამა, სვეტოვანი დიაგრამა, გაფანტულობის დიაგრამა, წრიული დიაგრამა, ჰისტოგრამა)..)

```python
import numpy as np
import matplotlib.pyplot as plt
df2 = pd.read_csv("Earthquakes_database.csv")
df2[["Month", "Day", "Year"]] = df2["Date"].str.split('/',2, expand=True)
df2
```

```python

df2['Year'] = pd.to_numeric(df2['Year'], errors='coerce')
plt.plot(df2['Year'], df2['Magnitude'], 'r--', lw=2)
plt.xlabel('Year')
plt.ylabel('Magnitude')
plt.title('Magnitude in Year')
plt.show()
```

```python
plt.scatter(df2['Year'], df2['Magnitude'], s=20, marker='^', c='Blue')
#plt.scatter(df2['Year'], df2['Depth'], s=20, marker='*', c='Red')

plt.show()
```

```python
df3 = df2.drop_duplicates(subset=['Type'])
x = df3['Type']
y = df3['Magnitude']
plt.bar(x, y, color='y', width=0.5, edgecolor='g', lw=3, ls='--')
plt.show()
```

```python
plt.hist(df2['Magnitude'], bins=15)
plt.xticks(np.arange(0,45, 5))
plt.show()
```
